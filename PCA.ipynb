```python
def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final
```

```python
numeric=df1._get_numeric_data()
df_without_outliers = remove_outlier_IQR(numeric)
df_without_outliers
```

```python
column_means = numeric.mean()
df2 = df_without_outliers.fillna(column_means)
```
