## Reading the dataframe

## Make the column names lower case

```python
def lower_case_column_names(hk_df):
    hk_df.columns=[i.lower() for i in hk_df.columns]
    return hk_df
```

### Deleting columns

```python
threshold_percent=0.1
print("Total records before drop",len(df1.columns))

df1_temp=df1.copy()
drop_threshold=math.floor(len(df1)*threshold_percent)
df1_temp.dropna(axis=1,thresh=drop_threshold,inplace=True)
len_of_temp=len(df1_temp.columns)
make_change=input(f"New file will contain {len_of_temp} records, commit? (y/n)")
if make_change.lower()=="y":
    df1.dropna(axis=1,thresh=drop_threshold,inplace=True)
else:
    print("Changes not implemented")
```

```python
threshold_percent=0.1
print("Total records before drop",len(df1))

df1_temp=df1.copy()
drop_threshold=math.floor(len(df1)*threshold_percent)
df1_temp.dropna(axis=0,thresh=drop_threshold,inplace=True)
len_of_temp=len(df1_temp)
make_change=input(f"New file will contain {len_of_temp} records, commit? (y/n)")
if make_change.lower()=="y":
    df1.dropna(axis=0,thresh=drop_threshold,inplace=True)
else:
    print("Changes not implemented")
```
